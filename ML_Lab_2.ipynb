```python
# Bar-Plot
```

```python
student = { "Vismit":68, "Jeet":50,"Vatsal":16 }
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names,values)
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Bar Graph")
plt.show()
```

```python
plt.barh(names,values,color="r")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.title("Bar Graph")
plt.show()
```
